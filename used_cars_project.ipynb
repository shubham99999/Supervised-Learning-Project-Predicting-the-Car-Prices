```python
import numpy as np
import pandas as pd
from sklearn import model_selection
```

```python
X_Y_train = pd.read_csv('training_car_x_y_train.csv', delimiter=',')
X_test = pd.read_csv('test_car_x_test.csv', delimiter=',')
```

```python
# deleting the unnecessary features/columns from training data
del X_Y_train["dateCrawled"]
del X_Y_train["seller"]
del X_Y_train["offerType"]
del X_Y_train["lastSeen"]
del X_Y_train["nrOfPictures"]
del X_Y_train["postalCode"]
del X_Y_train["dateCreated"]
X_Y_train.head(9)
```

```python
# removing the same columns as the training data , from the test data
del X_test["dateCrawled"]
del X_test["seller"]
del X_test["offerType"]
del X_test["nrOfPictures"]
del X_test["postalCode"]
del X_test["dateCreated"]
```

```python
del X_test["lastSeen"]
```

```python
X_Y_train.isnull().sum()
```

```python
# Dropping the rows having Nan in gearbox

X_Y_train.dropna(subset=["gearbox"],inplace=True)  #Drop only if NaN in gearbox column in training data
X_Y_train.reset_index(drop=True,inplace=True)
X_Y_train
```

```python
# Some more nulls are still remaining
X_Y_train.isnull().sum()
```

```python
# Now filling the Nan values with the most occuring object of the respective column!

X_Y_train = X_Y_train.apply(lambda x:x.fillna(x.value_counts().index[0])) 
# to fill NaNs with the most frequent value of that respective column.
X_Y_train
```

```python
print(X_Y_train.shape)
print(X_test.shape)
X_Y_train.isnull().sum()
# all NaN now removed!
```

```python
# removing the outliers/ exceptionally high or low values
# as we are interested in normal/practical ranged values
X_Y_train = X_Y_train[
        (X_Y_train.yearOfRegistration <= 2018) 
      & (X_Y_train.yearOfRegistration >= 1950) 
      & (X_Y_train.price >= 100) 
      & (X_Y_train.price <= 150000) 
      & (X_Y_train.powerPS >= 10) 
      & (X_Y_train.powerPS <= 500)]

print(X_Y_train.shape)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
def plot_correlation_map( df ):
    corr = df.corr()
    _ , ax = plt.subplots( figsize =( 12 , 10 ) )
    cmap = sns.diverging_palette( 220 , 10 , as_cmap = True )
    _ = sns.heatmap(
        corr, 
        cmap = cmap,
        square=True, 
        cbar_kws={ 'shrink' : .9 }, 
        ax=ax, 
        annot = True, 
        annot_kws = { 'fontsize' : 12 }
    )

plot_correlation_map(X_Y_train)
X_Y_train.corr()
```

# X_train_l and Y_train_l are the actual training data and target respectively

```python
X_train_l = X_Y_train.iloc[:,:X_Y_train.shape[1]-1]
Y_train_l = X_Y_train.iloc[:,-1]
print(Y_train_l)
X_train_l
```

```python

from sklearn import preprocessing
X_test=preprocessing.scale(X_test)
X_train_l = preprocessing.scale(X_train_l)
X_train_l.shape
X_train_l
```

```python
print(X_test.shape)
X_test
```

```python
from sklearn import model_selection
x_train,x_test,y_train,y_test= model_selection.train_test_split(X_train_l,Y_train_l,test_size=0.2)
print(x_train)
y_train
```

```python
from sklearn import linear_model
alg1=linear_model.LinearRegression()
alg1.fit(x_train,y_train)
y_test_pred= alg1.predict(x_test)
```

```python
# r2_score on sample train test for checking score
from sklearn.metrics import r2_score
r2_score(y_test,y_test_pred)
```

```python
from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor(random_state=0)
regressor.fit(x_train,y_train)
```

```python
# score for sample train test from decision tree
regressor.score(x_test,y_test)
```

```python
regressor.predict(x_test)
```
